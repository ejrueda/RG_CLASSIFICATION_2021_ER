```python
import pandas as pd
import numpy as np
```

Reference: 	Gerdes SY, et al (2003). Experimental determination and system level analysis of essential genes in Escherichia coli MG1655. J Bacteriol, 185:5673-84.['Supplemental Materials']

```python
data_essential_genes = pd.read_csv("../data/DEG_essential_genes_Ecoli_MG1655I.csv")
e_genes = data_essential_genes[" Gene Name"].values
data_essential_genes.head()
```

```python
ess_genes = []
for g in e_genes:
    for k in g.strip().split("/"):
        ess_genes.append(k)
```

```python
len(np.unique(ess_genes))
```

```python
candidates = pd.read_csv("../data/ICCSA_cantidates_06_03_20_v1.csv", index_col=0)
print("number of candidate genes:", len(candidates.index.unique()))
print(candidates.shape)
candidates.head()
```

```python
new_candidates = candidates.reindex(ess_genes).dropna()
new_candidates = new_candidates[~new_candidates.index.duplicated()]
new_candidates
```
